```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split

%matplotlib inline

df_ltp = pd.read_csv('/content/LTP.csv', encoding='latin-1')

df_ltp.head(10)
```

```python
rm_cols = ['laptop_ID']
df_ltp.drop(rm_cols, axis=1, inplace=True)

df_ltp.head(10)
```

```python
df_ltp.info()
```

```python
df_ltp.isna().sum()
```

```python
missing_values = df_ltp.isna().sum()
missing_percent = (missing_values / len(df_ltp)) * 100
print("Missing Values Percentage:\n", missing_percent)

for col in df_ltp.columns:
    if missing_percent[col] >= 30:
        # Drop the column if missing values are more than 30%
        df_ltp.drop(col, axis=1, inplace=True)
    else:
        # Jika missing values lebih dari 30%, sebaiknya dihapus, jika tidak gunakan median
        for col in df_ltp.select_dtypes(include=np.number).columns:
            df_ltp[col].fillna(df_ltp[col].median(), inplace=True)
```

```python
df_ltp.drop_duplicates(inplace=True)
df_ltp.duplicated().sum()
```

```python
df_ltp.info()
```

```python
df_ltp['Weight'] = df_ltp['Weight'].astype(str).str.replace('.', '').str.replace('kg', '')
df_ltp['Weight'] = df_ltp['Weight'].str.strip().str.replace('[^0-9\.]', '', regex=True)
df_ltp['Weight'] = df_ltp['Weight'].replace('', np.nan)
df_ltp['Weight'].fillna(0, inplace=True)
df_ltp['Weight'] = df_ltp['Weight'].astype(float)

df_ltp.info()
```

```python
df_ltp[num].hist(bins=30, figsize=(12, 6))
plt.show()
```

```python
df_ltp = df_ltp[(df_ltp['Inches'] > 0) & (df_ltp['Price_euros'] > 0)]
df_ltp['Inches'] = np.log(df_ltp['Inches'])
df_ltp['Price_euros'] = np.log(df_ltp['Price_euros'])

df_ltp.replace([np.inf, -np.inf], np.nan, inplace=True)

df_ltp.dropna(subset=['Inches', 'Price_euros'], inplace=True)

sns.boxplot(data=df_ltp['Inches'])
plt.title('Inches Boxplot')

plt.show()

sns.boxplot(data=df_ltp['Price_euros'])
plt.title('Price_euros Boxplot')

plt.show()
```

```python
df_ltp[num].hist(bins=30, figsize=(12, 6))
plt.show()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

df_ltp = remove_outliers_iqr(df_ltp, 'Inches')
df_ltp = remove_outliers_iqr(df_ltp, 'Price_euros')

sns.boxplot(data=df_ltp['Inches'])
plt.title('Inches Boxplot')
plt.show()

sns.boxplot(data=df_ltp['Price_euros'])
plt.title('Price_euros Boxplot')
plt.show()
```

```python
x = df_ltp.drop('Price_euros', axis=1)
y = df_ltp['Price_euros']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = x_train.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])
```
