```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split

%matplotlib inline

df_ltp = pd.read_csv('/content/LTP.csv', encoding='latin-1')

df_ltp.head(10)
```

```python
rm_cols = ['laptop_ID']
df_ltp.drop(rm_cols, axis=1, inplace=True)

df_ltp.head(10)
```

```python
df_ltp.drop_duplicates(inplace=True)
df_ltp.duplicated().sum()
```

```python
fix_cols = ['Company', 'Product', 'TypeName', 'ScreenResolution', 'Cpu', 'Ram', 'Memory', 'Gpu', 'OpSys', 'Weight']
for col in fix_cols:
  df_ltp[col] = df_ltp[col].astype('string')

df_ltp.info()
```

```python
df_ltp['Inches'] = np.log(df_ltp['Inches'])
df_ltp['Price_euros'] = np.log(df_ltp['Price_euros'])

df_ltp.replace([np.inf, -np.inf], np.nan, inplace=True)

df_ltp.dropna(subset=['Inches', 'Price_euros'], inplace=True)

sns.boxplot(data=df_ltp['Inches'])
plt.title('Inches Boxplot')

plt.show()

sns.boxplot(data=df_ltp['Price_euros'])
plt.title('Price_euros Boxplot')

plt.show()
```

```python
df_ltp[num].hist(bins=30, figsize=(12, 6))
plt.show()
```

```python
from scipy.stats.mstats import winsorize

# Winsorize data terlebih dahulu
df_ltp['Inches'] = winsorize(df_ltp['Inches'])  # Sesuaikan limits jika perlu
df_ltp['Price_euros'] = winsorize(df_ltp['Price_euros'])

# Kemudian terapkan IQR method
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR  # Sesuaikan pengali IQR jika perlu
    lower_bound = Q1 - 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

df_ltp = remove_outliers_iqr(df_ltp, 'Inches')
df_ltp = remove_outliers_iqr(df_ltp, 'Price_euros')

# Visualisasikan kembali dengan boxplot
sns.boxplot(data=df_ltp['Inches'])
plt.title('Inches Boxplot')
plt.show()

sns.boxplot(data=df_ltp['Price_euros'])
plt.title('Price_euros Boxplot')
plt.show()
```

```python
x = df_ltp.drop('Price_euros', axis=1)
y = df_ltp['Price_euros']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
```

```python
x_train['Inches'].hist(bins=20, figsize=(12,6))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train['Inches'] = scaler.fit_transform(x_train['Inches'].values.reshape(-1, 1))
x_test['Inches'] = scaler.transform(x_test['Inches'].values.reshape(-1, 1))

x_train['Inches'].hist(bins=20, figsize=(12, 6))
plt.show()
```
